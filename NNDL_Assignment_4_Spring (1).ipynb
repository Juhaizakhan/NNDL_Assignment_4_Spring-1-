```python
import pandas as pd
import matplotlib.pyplot as plt

# a) Read the provided CSV file
data = pd.read_csv("data.csv")

# c) Show basic statistical description
description = data.describe()

# d) Check for null values and replace them with mean
data.fillna(data.mean(), inplace=True)

# e) Aggregate data using min, max, count, mean
aggregated_data = data.agg({'Calories': ['min', 'max', 'count', 'mean'], 'Duration': ['min', 'max', 'count', 'mean']})

# f) Filter rows with calories values between 500 and 1000
filtered_data_1 = data.query('500 <= Calories <= 1000')

# g) Filter rows with calories > 500 and pulse < 100
filtered_data_2 = data.query('Calories > 500 & Pulse < 100')

# h) Create new dataframe without "Maxpulse" column
df_modified = data.drop(columns=['Maxpulse'])

# i) Delete "Maxpulse" column from main dataframe
data.drop(columns=['Maxpulse'], inplace=True)

# j) Convert datatype of "Calories" column to int
data['Calories'] = data['Calories'].astype(int)

# k) Create scatter plot for Duration and Calories
plt.scatter(data['Duration'], data['Calories'])
plt.xlabel('Duration')
plt.ylabel('Calories')
plt.title('Scatter Plot of Duration vs Calories')
plt.show()
```

```python
# 2) LINEAR REGRESSION

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# a) Read the provided CSV file
df = pd.read_csv("Salary_Data (2).csv")

# b) Split data into train and test partitions
X = df['YearsExperience'].values.reshape(-1, 1)
y = df['Salary'].values.reshape(-1, 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

# c) Train and predict the model
model = LinearRegression()
model.fit(X_train, y_train)
predictions_train = model.predict(X_train)
predictions_test = model.predict(X_test)

# d) Calculate mean squared error
mse_train = mean_squared_error(y_train, predictions_train)
mse_test = mean_squared_error(y_test, predictions_test)

# e) Visualize both train and test data
plt.scatter(X_train, y_train, color='blue', label='Train Data')
plt.scatter(X_test, y_test, color='red', label='Test Data')
plt.plot(X_train, predictions_train, color='green', label='Train Prediction')
plt.plot(X_test, predictions_test, color='orange', label='Test Prediction')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Linear Regression: Salary vs Years of Experience')
plt.legend()
plt.show()
```
